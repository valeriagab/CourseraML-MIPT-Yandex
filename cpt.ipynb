```python
# необходимые библиотеки

import numpy as np
import pandas as pd 
import scipy.stats as sts
import matplotlib.pyplot as plt
import random
```

```python
# зададим случайную велечину
df=random.randint(60,80)
# определяем функцию распределения с параметром df
chi=sts.chi(df)
```

```python
# сгенерируем 1000 значений
x=chi.rvs(size=1000)
```

```python
#гистаграмма выборки и теоритеская плотность распределения
plt.hist(x,normed=True,color='C')# строим гистограмму выборки
x=np.linspace(0,10)# определяем область теоретичской  плотности распределения
pdf=chi.pdf(x) # функция теоретиской функции распределения
plt.plot(x,pdf,'r', label='theoretical pdf') # график теоретической функции распределения
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
#Построим гистограмму распределения выборочных средних 1000 выборок из 5 элементов. 
#Построим график плотности вероятности нормального распределения среднего согласно ЦПТ.
means= [] # создаем пустой массив

for i in range(1000):
    var=chi.rvs(5)
    means.append(np.mean(var))
plt.hist(means,15,normed=True, label='mean hist')
plt.plot(x, sts.norm.pdf(x, chi.mean(),np.sqrt(chi.var()/5)),'r', label='norm pdf')
plt.xlabel('x')
plt.ylabel('fraction of samples')
plt.legend(loc='best')
    
```

```python
#Повторим для средних 1000 выборок из 10 элементов. 
means= []
for i in range(1000):
    var=chi.rvs(10)
    means.append(np.mean(var))
x=np.linspace(6,10)
plt.hist(means,15,normed=True, label='mean hist')
plt.plot(x, sts.norm.pdf(x, chi.mean(),np.sqrt(chi.var()/10)),'r', label='norm pdf')
plt.xlabel('x')
plt.ylabel('fraction of samples')
plt.legend(loc='best')
```

```python
#Повторим для средних 1000 выборок из 50 элементов. 
means= []
for i in range(1000):
    var=chi.rvs(50)
    means.append(np.mean(var))
x=np.linspace(7,10)
plt.hist(means,15,normed=True, label='mean hist')
plt.plot(x, sts.norm.pdf(x,chi.mean(),np.sqrt(chi.var()/50)),'r', label='norm pdf')
plt.xlabel('x')
plt.ylabel('fraction of samples')
plt.legend(loc='best')
```
